```python
import pandas as pd
import numpy as np
```

```python
#importing libraries
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_predict
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import LearningRateScheduler
```

```python
#reading data
data = pd.read_csv('result_df.csv')
```

```python
data
```

```python
x = data.iloc[:-2, 1:]

y = data.iloc[-1, 1:]
```

```python
x_t = x.transpose()
x_t
```

```python
y
```

```python
y = y.values.flatten()
```

```python
y = y.reshape(-1)
y
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
number_iterations = 100
auc_scores_dl=[]
auc_scores_dl_woDR = []
auc_scores_pca = {Name_of_model: [] for Name_of_model in ['Random Forest', 'SVM', 'AdaBoost', 'Logistic Regression', 'KNN']}
auc_scores_tsne = {Name_of_model: [] for Name_of_model in ['Random Forest', 'SVM', 'AdaBoost', 'Logistic Regression', 'KNN']}
auc_scores_ml = {Name_of_model: [] for Name_of_model in ['Random Forest', 'SVM', 'AdaBoost', 'Logistic Regression', 'KNN']}
```

```python
for iteration in range(number_iterations):

    print(f"Iteration {iteration + 1}/{number_iterations}")

    x_train, x_test, y_train, y_test = train_test_split(x_t, y, test_size = 0.2, random_state = None)

    #Standardization
    scaler = StandardScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)

    y_train = y_train.astype(float)
    y_test = y_test.astype(float)

    #pca for dimension reduction in ML models
    pca = PCA(n_components=10)
    x_train_pca = pca.fit_transform(x_train)
    x_test_pca = pca.fit_transform(x_test)

    #t-SNE for dimension reduction in ML models
    tsne = TSNE(n_components = 2, random_state=31)
    x_train_tsne = tsne.fit_transform(x_train)
    x_test_tsne = tsne.fit_transform(x_test)

    #ML models on PCA
    classifiers_pca = [RandomForestClassifier(), SVC(probability=True), AdaBoostClassifier(), LogisticRegression(), KNeighborsClassifier()]
    classifier_names_pca = ['Random Forest', 'SVM', 'AdaBoost', 'Logistic Regression', 'KNN']

    for model, name in zip(classifiers_pca, classifier_names_pca):
        y_pred = cross_val_predict(model, x_train_pca, y_train, cv =5, method='predict_proba' )[:,1]
        y_pred = np.array(y_pred)
        fpr, tpr, _ = roc_curve(y_train, y_pred)
        auc_score = auc(fpr, tpr)
        auc_scores_pca[name].append(auc_score)
        print(f'{name} AUC with PCA: {auc_score:.2f}')

    # ML models on t-SNE
    classifiers_tsne = [RandomForestClassifier(), SVC(probability=True), AdaBoostClassifier(), LogisticRegression(), KNeighborsClassifier()]
    classifier_names_tsne = ['Random Forest', 'SVM', 'AdaBoost', 'Logistic Regression', 'KNN']

    for model, name in zip(classifiers_tsne, classifier_names_tsne):
        y_pred = cross_val_predict(model, x_train_tsne, y_train, cv = 5, method='predict_proba' )[:,1]
        y_pred = np.array(y_pred)
        fpr, tpr, _ = roc_curve(y_train, y_pred)
        auc_score=auc(fpr, tpr)
        auc_scores_tsne[name].append(auc_score)
        print(f'{name} AUC with t-SNE: {auc_score:.2f}')



    #DEEP Learning Models with autoencoder

    ae = keras.Sequential([
        layers.Input(shape =(x_train.shape[1],)),
        layers.Dense(128, activation ='relu'),
        layers.Dense(64,activation='relu'),
        layers.Dense(10,activation='relu'),
        layers.Dense(64,activation='relu'),
        layers.Dense(128,activation='relu'),
        layers.Dense(x_train.shape[1],activation='sigmoid')
    ])

    ae.compile(optimizer='adam',loss='mean_squared_error')
    ae.fit(x_train, x_train, epochs=100, batch_size =32, verbose=0)

    encoder=keras.Sequential(ae.layers[:-3])

    x_train_ae = encoder.predict(x_train)
    x_test_ae = encoder.predict(x_test)

    #training with data reduction using auto-encoder

    dl_model = keras.Sequential([
    layers.Input(shape=(x_train_ae.shape[1],)),
    layers.Dense(128,activation='relu'),
    layers.Dense(64,activation='relu'),
    layers.Dense(1,activation='sigmoid')
    ])

    dl_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    dl_model.fit(x_train_ae, y_train, epochs=10, batch_size=32, verbose=0)

    y_pred_dl = dl_model.predict(x_test_ae)
    y_pred_dl = np.array(y_pred_dl)
    auc_score_dl = roc_auc_score(y_test, y_pred_dl)
    auc_scores_dl.append(auc_score_dl)
    print(f'Deep Learning AUC with Autoencoder: {auc_score_dl:.2f}')


```

```python
for iteration in range(number_iterations):

    print(f"Iteration {iteration + 1}/{number_iterations}")

    x_train, x_test, y_train, y_test = train_test_split(x_t, y, test_size = 0.2, random_state = None)

    #Standardization
    scaler = StandardScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)

    y_train = y_train.astype(float)
    y_test = y_test.astype(float)



  # ML models without dimension reduction
    classifiers_ml = [RandomForestClassifier(), SVC(probability=True), AdaBoostClassifier(), LogisticRegression(), KNeighborsClassifier()]
    classifier_names_ml = ['Random Forest', 'SVM', 'AdaBoost', 'Logistic Regression', 'KNN']

    for model, name in zip(classifiers_ml, classifier_names_ml):
      y_pred = cross_val_predict(model, x_train, y_train, cv=5, method='predict_proba')[:,1]
      y_pred = np.array(y_pred)
      fpr, tpr, _ = roc_curve(y_train, y_pred)
      auc_score = auc(fpr,tpr)
      auc_scores_ml[name].append(auc_score)
      print(f'{name} AUC without dimension reduction: {auc_score: .2f}')


    #DL model without dimension reduction

    #DL model with Normalization for better stability and acceleration

    model1 = keras.Sequential([
        layers.Input(shape=(x_train.shape[1],)),
        layers.BatchNormalization(),
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dense(256, activation='relu'),
        layers.Dense(128, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])

    def lr_schedule(epoch):
      return 0.001 * 0.9 ** epoch

    optimizer = Adam(learning_rate=0.001)
    model1.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

    # Training the model with learning rate scheduler
    model1.fit(x_train, y_train, epochs=100, batch_size=32, verbose=0, callbacks=[LearningRateScheduler(lr_schedule)])

    y_pred_dl = model1.predict(x_test)
    y_pred_dl = np.array(y_pred_dl)
    auc_score_dl = roc_auc_score(y_test, y_pred_dl)
    auc_scores_dl_woDR.append(auc_score_dl)
    print(f'Deep Learning AUC without Dimension Reduction: {auc_score_dl:.2f}')
```

```python
# Calculate and print the average AUC scores
# Calculate average AUC scores for DL
avg_auc_score_dl = sum(auc_scores_dl) / len(auc_scores_dl)
# Print average AUC scores for DL
print("Average AUC Scores for DL:")
print(f"Deep Learning: {avg_auc_score_dl:.2f}")

for name in ['Random Forest', 'SVM', 'AdaBoost', 'Logistic Regression', 'KNN']:
    average_auc_pca = sum(auc_scores_pca[name]) / number_iterations
    average_auc_tsne = sum(auc_scores_tsne[name]) / number_iterations
    print(f'{name} Average AUC with PCA: {average_auc_pca:.2f}')
    print(f'{name} Average AUC with t-SNE: {average_auc_tsne:.2f}')

# Calculate and print the average AUC score for the DL model without Dimension Reduction
average_auc_dl = sum(auc_scores_dl_woDR) / number_iterations
print(f'Deep Learning Average AUC without Dimension Reduction over {number_iterations} iterations: {average_auc_dl:.2f}')

# Calculate and print the average AUC scores for ML models without dimension reduction
for name in ['Random Forest', 'SVM', 'AdaBoost', 'Logistic Regression', 'KNN']:
    average_auc_ml = sum(auc_scores_ml[name]) / number_iterations
    print(f'{name} Average AUC without dimension reduction: {average_auc_ml:.2f}')
```
